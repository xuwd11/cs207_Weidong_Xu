See [Random sampling](https://docs.scipy.org/doc/numpy-1.13.0/reference/routines.random.html) for a bunch of `numpy` functions that draw random samples for you.  See [Statistical functions](https://docs.scipy.org/doc/scipy/reference/stats.html) for an alternative to `numpy`.  Plot a histogram of the data you generated.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Q1
def f1(x):
    return x/3

def f2(x):
    return np.sin(x)

def f3(x):
    return np.cos(x)

x = np.linspace(-np.pi, np.pi)
plt.plot(x, f1(x), label='$y=x/3$');
plt.plot(x, f2(x), label='$y=sin(x)$');
plt.plot(x, f3(x), label='$y=cos(x)$');
plt.xlabel('$x$');
plt.ylabel('$y$');
plt.legend();
plt.show()
```
